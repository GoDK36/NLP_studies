## 단어 빈도수 세기

```python
shin_raw = []
for talk in Shin:
    raw_talk = re.sub(' 회원님 : ', '', talk)
    shin_raw.append(raw_talk)
print(shin_raw[:10])

mins_raw = []
for talk in Mins:
    raw_talk = re.sub(' 주인공😘😙 : ', '', talk)
    mins_raw.append(raw_talk)
print(mins_raw[:10])
```

```python
import os, re, csv
import pandas as pd
from eunjeon import Mecab
from tqdm import tqdm
import gensim
import multiprocessing
from kiwipiepy import Kiwi
from soyspacing.countbase import CountSpace
```

```python
model = CountSpace()
model.load_model(r"E:\Develop\python\NLP\howls_nlp\Chatbot\자료\models\soy_big_space_model", json_format=False)

space_corrected_talk_s = []
for talk in shin_raw:
    # 띄어쓰기 시작
    corrected = model.correct(talk)
    space_corrected_talk_s.append(corrected)

spaced_talk_shin = []
for quest, tag in space_corrected_talk_s:
    if type(quest) == str:
        spaced_talk_shin.append(quest)
    else:
        continue

print(spaced_talk_shin)
```

```python
space_corrected_talk_m = []
for talk in mins_raw:
    # 띄어쓰기 시작
    corrected = model.correct(talk)
    space_corrected_talk_m.append(corrected)

spaced_talk_mins = []
for quest, tag in space_corrected_talk_m:
    if type(quest) == str:
        spaced_talk_mins.append(quest)
    else:
        continue

print(spaced_talk_mins)
```

```python
# Lemmatize

mecab = Mecab()

# 불용어 제거
file = open(r'E:\Develop\python\NLP\howls_nlp\Chatbot\자료\stop_words.txt','r', encoding = 'utf8')
lines = file.readlines()
file.close()
stop_words = lines[0].split()

def delete_stopwords(norm_content, test=False):
    noun_content = []
    for x in norm_content:
        noun_content.append(mecab.nouns(x))
    
    ### 추출된 명사들을 묶어 merge_content 에 담는다
    merge_content = []
    for noun in noun_content:
        txt = ' '.join(noun)
        merge_content.append(txt)
    
    ### 불용어 리스트에 없는 단어들만 추가, 명사들로 이뤄진 리스트들
    texts = [[word for word in document.split() if word not in stop_words]
            for document in merge_content]
    if test == True:
        return texts[0]
    return texts

def del_one(word_list):
    res = []
    for word in word_list:
        if len(word) > 1:
            res.append(word)
        else:
            continue
    return res

lemma_shin = delete_stopwords(spaced_talk_shin)
lemma_shin = sum(lemma_shin, [])
lemma_shin = del_one(lemma_shin)
lemma_mins = delete_stopwords(spaced_talk_mins)
lemma_mins = sum(lemma_mins, [])
lemma_mins = del_one(lemma_mins)

print(lemma_mins[:10])
print(lemma_shin[:10])
```

```python
#내용어 리스트 생성
shin_word_freq_Dict = {}
for lemma in lemma_shin:
    if lemma in shin_word_freq_Dict:
        shin_word_freq_Dict[lemma] += 1
    else:
        shin_word_freq_Dict[lemma] = 1

mins_word_freq_Dict = {}
for lemma in lemma_mins:
    if lemma in mins_word_freq_Dict:
        mins_word_freq_Dict[lemma] += 1
    else:
        mins_word_freq_Dict[lemma] = 1
print(mins_word_freq_Dict)
```

```python
import re


with open(r"E:\Develop\python\NLP\Corpus\KakaoTalkChats_고아.txt", encoding='utf-8') as f:
    conv = f.readlines()

del conv[0:6]
print(conv[:10])
```

```python
def kakao_preprocess(rsc, user):
    
    with open(rsc, encoding='utf-8') as f:
        conv = f.readlines()

    del conv[0:2]

    date_pat = re.compile(r"[0-9]{4}년 1?[0-9]월 [1-3]?[0-9]일 오[후전] [0-9]{1,2}:[0-9]{1,2}")

    talks = []
    for talk in conv:
        talk = re.sub(date_pat, '', talk)
        talk = re.sub('\n', '', talk)
        talk = re.sub(',', '', talk)
        if talk == '':
            pass
        elif user + ' : ' in talk:
            talks.append(talk)

    return talks
```
